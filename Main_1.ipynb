# Practice with 2 Variables

```python
import matplotlib.pyplot as plt

x = data["danceability"]
y = data["energy"]

plt.scatter(x, y)
plt.show()
```

```python
data1 = list(zip(x, y))
```

```python
from sklearn.cluster import KMeans

inertias = []

for i in range(1,12):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data1)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,12), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(data1)

plt.scatter(x, y, c=kmeans.labels_)
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=6)
kmeans.fit(data1)

plt.scatter(x, y, c=kmeans.labels_)
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
from sklearn.metrics import davies_bouldin_score
```

```python
from sklearn.datasets import make_blobs
```

```python
def createKMeans(k, data):
    data1 = list(zip(data["danceability"], data["energy"], data["loudness"], data["speechiness"], 
                     data["acousticness"], data["instrumentalness"], data["liveness"], data["valence"], data["tempo"]))
    
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data1)
    
    return kmeans
```

```python
X, y = make_blobs(n_samples=113999, centers=6, n_features=9, random_state=0)
```

```python
kMeans1 = createKMeans(6, data2).fit(X)
```

```python
X.shape
```

```python
davies_bouldin_score(X, kMeans1.labels_)
```

```python
def elbow_method(data):
    DB_scores = []
    
    for i in range(2, 10):
        X, y = make_blobs(n_samples=113999, centers=i, n_features=9, random_state=0)
        kmeans = createKMeans(i, data2).fit(X)
        DB_scores.append(davies_bouldin_score(X, kmeans.labels_))

    plt.plot(range(2,10), DB_scores, marker='o')
    plt.title('Elbow method')
    plt.xlabel('Number of clusters')
    plt.ylabel('Davies Bouldin Score')
    plt.show()
```

```python
elbow_method(data2)
```

```python
# Load dataset into a dataframe
df = pd.read_csv("dataset.csv")
df = df.drop("Unnamed: 0", axis="columns").drop_duplicates()
df = df.drop_duplicates(subset=['track_id'], keep = 'first')
df.drop(data[data["popularity"]==0].index, inplace=True)
```

```python
# Normalize dataset so all values lie within 0 and 1
from sklearn import preprocessing
scaler = preprocessing.MinMaxScaler()
df[chosen] = scaler.fit_transform(df[chosen])
```

```python
# Variables to Display
display = ["artists", "track_name"]
```

```python
chosen2 = ["liveness", "key", "valence",
          "speechiness", "acousticness", "instrumentalness"]
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(n_neighbors=10)
model.fit(df[chosen2])
```

```python
# Song Finder
df.loc[df['track_name']=='Biking']
```

```python
songs = model.kneighbors([df[chosen2].loc[103909]])[1][0]
```
